```python
#Series is a column
# A Pandas Series is like a column in a table.
# It is a one-dimensional array holding data of any type.

# Labels
# If nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc.
# This label can be used to access a specified value.

lst = [1,2,3,4,5,6]
p_series = pd.Series(lst)

#access each row by the index or label by default it is 0,1....

first_element = p_series[0]

#custom labels
lst = [1,2,3]
c_lbl = ["x","y","z"]
p_series = pd.Series(lst,index = c_lbl)
#same output
print(p_series[1])
print(p_series["x"])

#reindex
lst = [1,2,3]
c_lbl = ["x",2,"z"]
p_series = pd.Series(lst)
print(p_series)
p_series = p_series.reindex(c_lbl)
#if you reindex and previously and those all are new it will fill Nan values for those
print(p_series)

#series from a dict
dict = { "x":1,"y":2,"z":3}
p_series = pd.Series(dict)
print(p_series)
print(p_series[2])
```

```python
#A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

# creating a pandas dataframe
my_dataset = {
    "Emp_id" : [1,2,3,4,5],
    "Emp_name" : ["Abhi","Johanna","Sohail","Saurabh","Antonia"]
}

df = pd.DataFrame(my_dataset)
df.head(5)

#locate row
print(df.loc[[0,1]])

#select column
print(df[["Emp_id","Emp_name"]])

#reindex act similar to series
lbl = ["x","y","z","a",2]
df = df.reindex(lbl)
print(df)

#analyze
print("head and tail")
df.head()
df.tail()
```

```python
import pandas as pd

# Sample DataFrames
data1 = {'key1': ['A', 'B', 'C', 'D'],
         'value1': [1, 2, 3, 4]}
data2 = {'key': ['B', 'C', 'E', 'F'],
         'value2': [5, 6, 7, 8]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Inner Join
result = df1.merge(df2,  left_on='key1', right_on='key', how='inner')
print(result)
```

```python
import pandas as pd

# Sample DataFrames
data1 = {'A': ['A1', 'A2', 'A3'],
         'B': ['B1', 'B2', 'B3']}
data2 = {'C': ['A4', 'A5', 'A6'],
         'D': ['B4', 'B5', 'B6']}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

df2.rename(columns={"C":"A","D":"B"},inplace=True)

# Union of two DataFrames
union_result = pd.concat([df1, df2],axis=1)

print(union_result)
```

```python
import pandas as pd

# Sample DataFrames with different column names
data1 = {'A': ['A1', 'A2', 'A3'],
         'B': ['B1', 'B2', 'B3']}
data2 = {'X': ['A4', 'A5', 'A6'],
         'Y': ['B4', 'B5', 'B6']}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Concatenate vertically with different column names
result = pd.concat([df1, df2], ignore_index=True)

print(result)
```
