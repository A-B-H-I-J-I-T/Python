```python
#Series is a column
# A Pandas Series is like a column in a table.
# It is a one-dimensional array holding data of any type.

# Labels
# If nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc.
# This label can be used to access a specified value.

lst = [1,2,3,4,5,6]
p_series = pd.Series(lst)

#access each row by the index or label by default it is 0,1....

first_element = p_series[0]

#custom labels
lst = [1,2,3]
c_lbl = ["x","y","z"]
p_series = pd.Series(lst,index = c_lbl)
#same output
print(p_series[1])
print(p_series["x"])

#reindex
lst = [1,2,3]
c_lbl = ["x",2,"z"]
p_series = pd.Series(lst)
print(p_series)
p_series = p_series.reindex(c_lbl)
#if you reindex and previously and those all are new it will fill Nan values for those
print(p_series)

#series from a dict
dict = { "x":1,"y":2,"z":3}
p_series = pd.Series(dict)
print(p_series)
print(p_series[2])
```

```python
#A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

# creating a pandas dataframe
my_dataset = {
    "Emp_id" : [1,2,3,4,5],
    "Emp_name" : ["Abhi","Johanna","Sohail","Saurabh","Antonia"]
}

df = pd.DataFrame(my_dataset)
df.head(5)

#locate row
print(df.loc[[0,1]])

#select column
print(df[["Emp_id","Emp_name"]])

#reindex act similar to series
lbl = ["x","y","z","a",2]
df = df.reindex(lbl)
print(df)

#analyze
print("head and tail")
df.head()
df.tail()
```
